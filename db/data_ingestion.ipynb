# Data Ingestion
Ce code permet de nettoyer les données des stations afin de faciliter l'ingestion avec les colonnes demandées en entrée.

```python
# import
import os
import pandas as pd 
import geopandas as gpd 
from datetime import datetime
```

```python
# read stations geojson
stations_gdf = gpd.read_file("../data/station_abidjan_v1.geojson")

display(stations_gdf)
```

```python
# change id, ville_id & created_at column with numbers
stations_gdf["id"] = range(1, len(stations_gdf)+1)
stations_gdf["ville_id"] = 1
stations_gdf["created_at"] = datetime.now()
```

```python
# rename columns
matches_cols = {
    "name": "nom",
    "operator": "operateur",
    "address": "adresse"
}

stations_gdf = stations_gdf.rename(columns=matches_cols)
```

```python
# clean shop dtypes
display(stations_gdf.dtypes)

stations_gdf["shop"] = stations_gdf["shop"].apply(
    lambda shop : False if shop in [None, "fuel", "kiosk", "no"] else True
)
stations_gdf["shop"] = stations_gdf["shop"].astype(bool)

display(set(stations_gdf["shop"]))
```

```python
display(stations_gdf)
```

```python
# export
stations_gdf.to_file("../data/shared/stations.geojson")
stations_gdf.to_file("../data/shared/stations.shp")
```
