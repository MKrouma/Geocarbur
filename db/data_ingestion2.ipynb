# Ingestion de données

```python
import os 
import pandas
import geopandas
from datetime import datetime
```

```python
# lire le fichier geojson des stations
gdf = geopandas.read_file("../data/station_abidjan_v1.geojson")
```

```python
# afficher les données
display(gdf)
display(gdf.sample())
```

```python
# exploitation
gdf.describe()
```

```python
# renommer les colonnes
gdf_clean = gdf.rename(columns={
    "name": "nom",
    "operator": "operateur",
    "address": "adresse",
})

display(gdf.columns)
display(gdf_clean)
```

```python
# ajouter les colonnes ville_id et created_at
gdf_clean["ville_id"] = 1
gdf_clean["created_at"] = datetime.now()

display(gdf_clean.sample())
```

```python
# nettoyer id et shop
gdf_clean["id"] = range(1, 324)
gdf_clean["shop"] = gdf_clean["shop"].apply(
    lambda shop: True if shop=="yes" else False
)

display(gdf_clean.dtypes)
display(gdf_clean.sample())
```

```python
# exporter gdf clean
gdf_clean.to_file("../data/resultat/stations.shp")
gdf_clean.to_file("../data/resultat/stations.geojson")
```
